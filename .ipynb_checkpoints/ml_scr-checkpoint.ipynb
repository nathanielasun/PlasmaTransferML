```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
import numpy as np
import random
```

```python
def count_parameters(model, show_all=False):
    total_params = 0
    if (show_all):
        i=1
        for p in model.parameters():
            print("Layer ", i, " ", p.numel())
            total_params+=p.numel()
            i+=1
    else:
        total_params = sum(p.numel() for p in model.parameters())
        
    print("total params: ", total_params)
    return total_params
```

```python
def set_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)  # For GPU (if you're using one)
    torch.cuda.manual_seed_all(seed)  # If using multi-GPU
    np.random.seed(seed)  # For numpy random numbers
    random.seed(seed)  # For Python's built-in random module
    torch.backends.cudnn.deterministic = True  # Make cuDNN deterministic
    torch.backends.cudnn.benchmark = False  # Ensure consistency
```

```python
def initialize_weights(module):
    """
    Generic initialization routine that applies Xavier initialization
    to weights and sets biases to zero (if applicable).
    """
    if isinstance(module, (nn.Linear, nn.Conv2d)):  # For Linear and Conv layers
        torch.nn.init.xavier_uniform_(module.weight)
        if module.bias is not None:
            torch.nn.init.zeros_(module.bias)
    elif isinstance(module, nn.LSTM):  # For LSTM layers
        for name, param in module.named_parameters():
            if 'weight_ih' in name or 'weight_hh' in name:
                torch.nn.init.xavier_uniform_(param)
            elif 'bias' in name:
                torch.nn.init.zeros_(param)
```

```python
class CNN_DP(nn.Module): #CNN module for EAST DP as outlined by Guo et al. 2021
    def __init__(self):
        super(CNN_DP, self).__init__()
        self.conv1_1 = nn.Conv1d(14, 64, 3)
        self.conv1_2 = nn.Conv1d(64, 64, 3)
        self.conv2_1 = nn.Conv1d(64, 128, 3)
        self.conv2_2 = nn.Conv1d(128, 128, 3)
        self.conv3_1 = nn.Conv1d(128, 256, 5)
        self.conv3_2 = nn.Conv1d(256, 64, 1)
        self.conv3_3 = nn.Conv1d(64, 1, 1)
        self.pool2x2 = nn.MaxPool1d(2, 2)
        self.pool3x3 = nn.MaxPool1d(3, 3)

    def forward(self, input):
        x1 = F.relu(self.conv1_1(input))
        x2 = F.relu(self.conv1_2(x1))
        x3 = self.pool2x2(x2)
        x4 = F.relu(self.conv2_1(x3))
        x5 = F.relu(self.conv2_2(x4))
        x6 = self.pool3x3(x5)
        x7 = F.relu(self.conv3_1(x6))
        x8 = F.relu(self.conv3_2(x7))
        x9 = F.sigmoid(self.conv3_3(x8))
        return x9
    #initial LR = 0.001, ADAM
```

```python
class LSTM_DP(nn.Module):
    def __init__(self):
        super(LSTM_DP, self).__init__()
        self.LSTM_1 = nn.LSTM(input_size=14, hidden_size=396, batch_first=True)
        self.LSTM_2 = nn.LSTM(input_size=396, hidden_size=101, batch_first=True)
        self.dense = nn.Linear(101, 1)
        
    def forward(self, input):
        x1 = F.tanh(self.LSTM_1(input))
        x2 = F.tanh(self.LSTM_2(x1))
        x2 = x2[:, -1, :]
        x3 = F.sigmoid(self.dense(x2))
        return x3
```
