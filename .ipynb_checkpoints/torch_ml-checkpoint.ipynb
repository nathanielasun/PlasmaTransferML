```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
import random
import numpy as np
```

```python
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

batch_size = 10

trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=8)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=batch_size,
                                         shuffle=False, num_workers=8)

classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
def initialize_weights(module):
    """
    Generic initialization routine that applies Xavier initialization
    to weights and sets biases to zero (if applicable).
    """
    if isinstance(module, (nn.Linear, nn.Conv2d)):  # For Linear and Conv layers
        torch.nn.init.xavier_uniform_(module.weight)
        if module.bias is not None:
            torch.nn.init.zeros_(module.bias)
    elif isinstance(module, nn.LSTM):  # For LSTM layers
        for name, param in module.named_parameters():
            if 'weight_ih' in name or 'weight_hh' in name:
                torch.nn.init.xavier_uniform_(param)
            elif 'bias' in name:
                torch.nn.init.zeros_(param)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# functions to show an image


def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()


# get some random training images
dataiter = iter(trainloader)
images, labels = next(dataiter)

# show images
imshow(torchvision.utils.make_grid(images))
# print labels
print(' '.join(f'{classes[labels[j]]:5s}' for j in range(batch_size)))
```

```python
def count_parameters(model):
    total_params = sum(p.numel() for p in model.parameters())
    return total_params
```

```python
def set_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)  # For GPU (if you're using one)
    torch.cuda.manual_seed_all(seed)  # If using multi-GPU
    np.random.seed(seed)  # For numpy random numbers
    random.seed(seed)  # For Python's built-in random module
    torch.backends.cudnn.deterministic = True  # Make cuDNN deterministic
    torch.backends.cudnn.benchmark = False  # Ensure consistency
```

```python
class CBAM(nn.Module):
    def __init__(self, in_channels, reduction=30):
        super(CBAM, self).__init__()
        self.channel_attention = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(in_channels, in_channels // reduction, kernel_size=1),
            nn.ReLU(),
            nn.Conv2d(in_channels // reduction, in_channels, kernel_size=1),
            nn.Sigmoid()
        )
        self.spatial_attention = nn.Sequential(
            nn.Conv2d(2, 1, kernel_size=7, padding=3),
            nn.Sigmoid()
        )

    def forward(self, x):
        # Channel Attention
        ca = self.channel_attention(x) * x
        
        # Spatial Attention
        max_out, _ = torch.max(ca, dim=1, keepdim=True)
        avg_out = torch.mean(ca, dim=1, keepdim=True)
        sa = torch.cat([max_out, avg_out], dim=1)
        sa = self.spatial_attention(sa) * ca
        return sa
```

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1_1 = nn.Conv2d(3, 30, 1, padding=0)
        self.conv1_3 = nn.Conv2d(3, 30, 3, padding=1)
        self.conv1_5 = nn.Conv2d(3, 30, 5, padding=2)
        self.conv1_7 = nn.Conv2d(3, 30, 7, padding=3)
        self.conv2_1 = nn.Conv2d(30, 30, 1, padding=0)
        self.conv2_3 = nn.Conv2d(30, 30, 3, padding=1)
        self.conv2_5 = nn.Conv2d(30, 30, 5, padding=2)
        self.conv2_7 = nn.Conv2d(30, 30, 7, padding=3)
        self.fc1 = nn.Linear(4*30*32*32,  200)
        self.fc2 = nn.Linear(200, 10)

    def forward(self, input):
        x0 = F.relu(self.conv1_1(input))
        x1 = F.relu(self.conv1_3(input))
        x2 = F.relu(self.conv1_5(input))
        x3 = F.relu(self.conv1_7(input))
        x0 = F.relu(self.conv2_7(x0))
        x1 = F.relu(self.conv2_1(x1))
        x2 = F.relu(self.conv2_3(x2))
        x3 = F.relu(self.conv2_5(x3))
        x = torch.flatten(torch.cat((x0, x1, x2, x3), dim=1), 1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
device = torch.device("mps")
```

```python
set_seed(239001)
net = Net().to(device)
net.apply(initialize_weights)
```

```python
class DenseLayer(nn.Module):
    def __init__(self, in_channels, growth_rate, bottleneck=True):
        super(DenseLayer, self).__init__()
        self.bottleneck = bottleneck
        if bottleneck:
            self.bn1 = nn.BatchNorm2d(in_channels)
            self.conv1 = nn.Conv2d(in_channels, growth_rate * 4, kernel_size=1, bias=False)
        self.bn2 = nn.BatchNorm2d(growth_rate * 4 if bottleneck else in_channels)
        self.conv2 = nn.Conv2d(growth_rate * 4 if bottleneck else in_channels, growth_rate, kernel_size=3, padding=1, bias=False)

    def forward(self, x):
        if self.bottleneck:
            x = F.relu(self.bn1(x))
            x = self.conv1(x)
        out = F.relu(self.bn2(x))
        out = self.conv2(out)
        return torch.cat([x, out], dim=1)  # Concatenate input and output along the channel dimension


class DenseBlock(nn.Module):
    def __init__(self, num_layers, in_channels, growth_rate, bottleneck=True):
        super(DenseBlock, self).__init__()
        self.layers = nn.ModuleList()
        for i in range(num_layers):
            self.layers.append(DenseLayer(in_channels + i * growth_rate, growth_rate, bottleneck))

    def forward(self, x):
        for layer in self.layers:
            x = layer(x)
        return x


class TransitionLayer(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(TransitionLayer, self).__init__()
        self.bn = nn.BatchNorm2d(in_channels)
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size=1, bias=False)
        self.pool = nn.AvgPool2d(kernel_size=2, stride=2)

    def forward(self, x):
        x = F.relu(self.bn(x))
        x = self.conv(x)
        x = self.pool(x)
        return x


class DenseNet(nn.Module):
    def __init__(self, growth_rate=12, num_blocks=3, num_layers_per_block=4, num_classes=10, init_channels=24, bottleneck=True):
        super(DenseNet, self).__init__()
        self.growth_rate = growth_rate

        # Initial convolution
        self.conv1 = nn.Conv2d(3, init_channels, kernel_size=3, padding=1, bias=False)

        # Dense blocks and transition layers
        self.blocks = nn.ModuleList()
        num_channels = init_channels
        for i in range(num_blocks):
            # Add dense block
            self.blocks.append(DenseBlock(num_layers=num_layers_per_block, in_channels=num_channels, growth_rate=growth_rate, bottleneck=bottleneck))
            num_channels += num_layers_per_block * growth_rate
            # Add transition layer (except after the last block)
            if i != num_blocks - 1:
                self.blocks.append(TransitionLayer(in_channels=num_channels, out_channels=num_channels // 2))
                num_channels = num_channels // 2

        # Classification layer
        self.bn = nn.BatchNorm2d(num_channels)
        self.fc = nn.Linear(num_channels, num_classes)

    def forward(self, x):
        old_x = x
        if self.bottleneck:
            x = F.relu(self.bn1(x))
            x = self.conv1(x)
        x = F.relu(self.bn2(x))
        x = self.conv2(x)
        # Now concatenate old_x (all the old stuff) with the new feature maps
        return torch.cat([old_x, x], dim=1)


# Instantiate and test the model
model = DenseNet(growth_rate=12, num_blocks=3, num_layers_per_block=4, num_classes=10, init_channels=24).to(device)
print(model)
```

```python
count_parameters(model)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=.9)
```

```python
for epoch in range(20):
    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        inputs, labels = data
        inputs, labels = inputs.to(device), labels.to(device)
        indices = torch.randperm(inputs.size(0))
        inputs, labels = inputs[indices], labels[indices]
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        if i % 500 == 499:    # print every 2000 mini-batches
            print(f'[{epoch + 1}, {i + 1:5d}] loss: {running_loss / 500:.3f}')
            running_loss = 0.0

print('Finished Training')
```

```python
dataiter = iter(testloader)
images, labels = next(dataiter)

# print images
imshow(torchvision.utils.make_grid(images))
print('GroundTruth: ', ' '.join(f'{classes[labels[j]]:5s}' for j in range(8)))
```

```python
outputs = net(images.to(device))
_, predicted = torch.max(outputs, 1)

print('Predicted: ', ' '.join(f'{classes[predicted[j]]:5s}'
                              for j in range(8)))
```

```python
correct = 0
total = 0

with torch.no_grad():
    for data in testloader:
        images, labels = data
        images, labels = images.to(device), labels.to(device)
        # calculate outputs by running images through the network
        outputs = net(images)
        # the class with the highest energy is what we choose as prediction
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f'Accuracy of the network on the 10000 test images: {100 * correct // total} %')
```
